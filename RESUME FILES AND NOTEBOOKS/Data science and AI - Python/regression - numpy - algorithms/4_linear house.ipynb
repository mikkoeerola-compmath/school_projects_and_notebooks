```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Sample data: Size (m²) vs Price (thousands of €)
X = np.array([[30], [40], [50], [60], [70], [80], [90], [100]])  # feature
y = np.array([100, 150, 200, 220, 250, 280, 310, 340])           # target

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Predict using the same X values
y_pred = model.predict(X)

# Plot the results
plt.scatter(X, y, color='blue', label='Actual data')
plt.plot(X, y_pred, color='red', label='Regression line')
plt.xlabel("Size (m²)")
plt.ylabel("Price (thousands €)")
plt.title("Simple Linear Regression")
plt.legend()
plt.grid(True)
plt.show()

# Print coefficients
print("Slope (coefficient):", model.coef_[0])
print("Intercept:", model.intercept_)
```

In a linear regression model, the slope (coefficient) represents how much the predicted value increases for each unit increase in the input;
for example, a slope of 3.27 means that for every additional square meter,
the predicted house price increases by €3,270.
The intercept, here 18.45, is the predicted value when the input (e.g., size) is 0 — it’s the starting point of the line on the y-axis.
Together, they form the regression equation:

Price=3.27⋅Size+18.45
So, for a 60 m² house, the model would predict a price of around €214,650.

Goal: Predict house price based on:
Size (in m²)
Number of rooms

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

# Input features: [size in m², number of rooms]
X = np.array([
    [50, 2],
    [60, 2],
    [70, 3],
    [80, 3],
    [90, 4],
    [100, 4],
    [110, 5],
    [120, 5]
])

#  Target: price in thousands of €
y = np.array([150, 180, 210, 240, 270, 300, 330, 360])

# Train the model
model = LinearRegression()
model.fit(X, y)

# Show coefficients
print("Coefficients:", model.coef_)       # one for each feature
print("Intercept:", model.intercept_)     # base price

# Predict a new example
new_house = np.array([[85, 3]])  # 85 m², 3 rooms
predicted_price = model.predict(new_house)
print("Predicted price for 85 m², 3 rooms:", predicted_price[0], "thousand €")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from mpl_toolkits.mplot3d import Axes3D  # For 3D plotting

# Data: [size, rooms] and price
X = np.array([
    [50, 2],
    [60, 2],
    [70, 3],
    [80, 3],
    [90, 4],
    [100, 4],
    [110, 5],
    [120, 5]
])
y = np.array([150, 180, 210, 240, 270, 300, 330, 360])

# Train the model
model = LinearRegression()
model.fit(X, y)

# Create 3D plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')

# Plot original data
ax.scatter(X[:, 0], X[:, 1], y, color='blue', label='Actual prices')

# Create a mesh grid for the surface
x_surf, y_surf = np.meshgrid(
    np.linspace(X[:, 0].min(), X[:, 0].max(), 10),
    np.linspace(X[:, 1].min(), X[:, 1].max(), 10)
)
z_surf = model.predict(np.c_[x_surf.ravel(), y_surf.ravel()]).reshape(x_surf.shape)

# Plot regression surface
ax.plot_surface(x_surf, y_surf, z_surf, color='red', alpha=0.5)

# Labels
ax.set_xlabel('Size (m²)')
ax.set_ylabel('Rooms')
ax.set_zlabel('Price (thousand €)')
ax.set_title('Multiple Linear Regression (3D)')
ax.legend()

plt.show()
```

This multiple linear regression example uses two input features — house size (in square meters) and number of rooms — to predict the price of a house. After training the model with sample data, a 3D plot is created to visualize both the actual data points and the regression surface. The blue dots in the plot represent the real house prices based on different combinations of size and room count, while the red surface illustrates the model’s predictions across that range. This visualization clearly shows how both features influence the predicted price and helps us understand the linear relationship in a multi-dimensional space.
