# Pandas example 1: Wine

The [Wine dataset from UCI machine learning repository](https://archive.ics.uci.edu/dataset/186/wine+quality) contains chemical analysis results from various Portugues wines, together with a numerical "quality" value. We shall have a look at the dataset and try to have a look at how "good" wines differ from "average" wines.

```python
import pandas as pd
df = pd.read_csv('winequality-white.csv',sep=';')
df
```

```python
df.columns
```

## General statistics

```python
df.describe()
```

## How many of different quality?

```python
df['quality'].value_counts()
```

## Good, bad, average
We consider a wine to be *average* if its quality is 5 or 6. We consider it *bad* if it's less than 5 and *good* if it's at least 7.

How many wines of different quality?

```python
df_bad = df[df['quality'] < 5]
df_bad.shape
```

```python
df_average = df[(df['quality'] == 5) | (df['quality'] == 6)]
df_average.shape
```

```python
df_good = df.query('quality >= 7')
df_good.shape
```

## Small subset of features for good vs. average vs. bad wines

Select a subset of features and compute their means across the different groups of wines to see if there are connections.

```python
df.columns
```

```python
features = ['fixed acidity', 'chlorides', 'sulphates', 'alcohol']
```

```python
df_good[features].mean()
```

```python
df_average[features].mean()
```

```python
df_bad[features].mean()
```

We can also use array operations

```python
df_good[features].mean() - df_bad[features].mean()
```
